```python
#Count duplicates
df.duplicated().sum()
```

```python
df["age"] = pd.to_numeric(df["age"], errors="coerce")
```

```python
#Identify impossible values:

#Age < 18 or > 60
df["impossiblr_age"] = ((df["age"]<18) | (df["age"]>60))
```

```python
#Salary < 15000 or > 300000
df["impossible_Salary"] = ((df["salary"]<15000)| (df["salary"]>30000))
```

```python
#Experience > age
df["experience_great_age"] = df["experience_years"] > df["age"]
```

```python
#Joining_date missing or invalid
df["joining_date_missing"] = df["joining_date"].isna()
```

```python
#2 DATA CLEANING (THE CORE OF THE PROJECT)

#Task 3: Clean names
df.dropna(subset= ["name"], inplace=True)
```

```python
#Remove leading/trailing spaces
df["name"] = df["name"].str.strip()
```

```python
#Fill missing names with "Unknown"
df["name"] = df["name"].fillna("unknown")
```

```python
#Fix duplicates (Ali Khan appears multiple times)
df.drop_duplicates(subset=["name"], inplace=True)
```

```python
#Task 4: Clean gender column

#Convert to lowercase
df["gender"] = df["gender"].str.lower()
```

```python
#Replace inconsistent values:

#m → Male
df["gender"] = df["gender"].replace({"m":"Male"})
```

```python
#f → Female
df["gender"] = df["gender"].replace({"f":"Female"})
```

```python
#empty → "Unknown"
df["gender"] = df["gender"].fillna("Unknown")
```

```python
#Task 5: Clean department & job_role

#Normalize job_role (e.g., data analyst, hr manager)
df["job_role"] = df["job_role"].str.lower()
df["job_role"] = df["job_role"].str.strip()
```

```python
#Replace missing department with "Unknown"
df["department"] = df["department"].fillna("unknown")
```

```python
#Convert department to Title Case
df["department"] = df["department"].str.title()
```

```python
#Task 6: Fix numerical issues

#Replace ages > 100 with NaN
df.loc[df["age"]>100,"age"] = np.nan
```

```python
#Fill NaN salaries with median by job_role
df["salary"] = df.groupby("job_role")["salary"].transform(lambda x: x.fillna(x.median()))
```

```python
#Fill NaN age with median age
df["age"] = df["age"].fillna(df["age"].median())
```

```python
#Check if experience > age → fix or replace
df.loc[df["experience_years"]> df["age"], "experience_years"] =np.nan
df["experience_years"] = df.groupby("job_role")["experience_years"].transform(lambda x: x.fillna(x.median()))
```

```python
#Task 7: Handle Missing Values

#Fill missing attrition as "No"
df["attrition"] = df["attrition"].fillna("No")
```

```python
#Fill missing joining_date with "Unknown" or drop (your choice)
df["joining_date"] = df["joining_date"].fillna("Unknown")
```

```python
#Task 8: Convert Data Types

#Convert joining_date → datetime
#df["joining_date"] = pd.to_datetime(df["joining_date"])
df["joining_date"] = pd.to_datetime(df["joining_date"], errors="coerce")
```

```python
#Task 9: Remove Duplicate Rows

#Drop duplicates using employee_id, name, job_role combination (your chosen keys)
df.drop_duplicates(subset=["employee_id","name","job_role"],inplace=True)
```

```python
#3️⃣ FEATURE ENGINEERING
#Task 10: Create new columns

#tenure_years = 2024 − joining year
df["joining_year"] = df["joining_date"].dt.year
df["joining_year"] = df["joining_year"].astype("Int64")
df["tenure_year"] = 2024 - df["joining_year"]
```

```python
"""salary_level:

Low (<50k)

Mid (50–100k)

High (>100k)"""

df["Salary_level"] = np.where(df["salary"]<50000, "low", np.where((df["salary"]>=50000) & (df["salary"]<100000), "Mid","High"))
```

```python
"""age_group:

Young (18–30)

Mid (31–45)"""

df["age_level"] = np.where((df["age"]>18)&(df["age"]<30),"Young", np.where((df["age"]>31)&(df["age"]<45),"Mid","Senoir"))
```

```python
#Task 11: Attrition Flag

#Convert Yes/No → 1/0 column
#attrition_flag
df["attrition_flag"] = df["attrition"].map({"Yes":1, "No":0}).fillna(0)
```

```python
#Task 12: Salary Adjustment

#Cap salaries at realistic boundaries (20k–250k)
df["salary"] = df["salary"].clip(lower=20000, upper=250000)
```

```python
#4 ANALYSIS & INSIGHTS
#Task 13: Department-Level Insights

#Employee count by department
df.groupby("department")["employee_id"].count().reset_index()
```

```python
#Task 16: Experience & Salary Analysis

#Correlation between experience and salary
correlation = df["experience_years"].corr(df["salary"])
print(correlation)
```

```python
#Does more experience = higher salary?
if correlation > 0.5:
    print("Yes, more experience generally lead to high salary")
elif correlation >0:
    print("there is weak chance of salary ")
else:
    print("there's no chance of salary")
```

```python
#Does tenure affect attrition?
tenure_affect = df["tenure_year"].corr(df["attrition_flag"])
print(tenure_affect)

if tenure_affect <-0.3:
    print("Yes, employee with longer tenure are less likley to leave")
elif tenure_affect >-0.3:
    print("employee with more tenure are more likely to leave")
else:
    print("depend on the employee")
```

```python
#Bar chart → Experience vs salary bucket
bins = [0, 30000, 60000, 90000, 120000, 200000]
labels = ["0–30k", "30–60k", "60–90k", "90–120k", "120k+"]

df["salary_bucket"] = pd.cut(df["salary"], bins=bins, labels=labels, include_lowest=True)
exp_vs_salary = df.groupby("salary_bucket")["experience_years"].mean()

plt.figure(figsize=(10,5))
plt.bar(exp_vs_salary.index, exp_vs_salary.values, edgecolor="black")
plt.title("Experience Vs Salary Bucket")
plt.xlabel("Salary Bucket")
plt.ylabel("Average Experience (Years)")
plt.xticks(rotation=45)
plt.show()
```
